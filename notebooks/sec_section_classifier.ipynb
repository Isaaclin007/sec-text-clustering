```python
def preformat_document(document):
    document = document.split('</Header>', 1)[-1]
    document = document.replace("    ", "tab4")
    return document
# main = lines.split('</Header>', 1)[-1]
```

```python
main = main.replace("    ", "tab4")
```

```python
def locate_section_header(document, header, window=15, large_window=100):
    
    # find matches for section header
    candidates = [(match.start(), match.end()) for match in re.finditer(header, document, re.IGNORECASE)]
    
    # calculate metrics
    metrics = []
    for cand in candidates:
        line_break_count = document[cand[0] - window: cand[1] + window].count('\n')
        global_line_break_count = document[cand[0] - large_window: cand[1] + large_window].count('\n')
        tab_count = document[cand[0] - window: cand[1] + window].count('tab4')
        is_caps = document[cand[0]:cand[1]].isupper()
        metrics.append({'num_line_breaks': line_break_count, 'is_caps': is_caps, 'num_tabs': tab_count, 'global_line_break_count': global_line_break_count})
#         print
    
    # evaluate metrics
    max_linebreaks = max(metrics, key=lambda d: d['num_line_breaks']).get('num_line_breaks', 1)
    scores = []
    for metric in metrics:
        scores.append(float(metric['num_line_breaks']) / max_linebreaks + metric['is_caps'] - 0.1 * metric['num_tabs'] - 1 *(metric['global_line_break_count'] > 10))
    
    print([m['global_line_break_count'] for m in metrics])
    print(scores)
    
    return candidates[scores.index(max(scores))]
    
start, end = locate_section_header(main, 'Risk Factors')
print(main[start-20:end+20])
```
